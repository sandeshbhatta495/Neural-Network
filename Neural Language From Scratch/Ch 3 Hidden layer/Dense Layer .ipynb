```python
import numpy as np
```

```python
class Layer_Dense:
    # Initialize the layer's parameters
    def __init__(self, n_inputs, n_neurons):
        # Weights matrix: (n_inputs, n_neurons) is more conventional for matrix multiplication
        # using the dot product in the forward pass.
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        # Biases vector: (1, n_neurons) ensures correct broadcasting during summation
        self.biases = np.zeros((1, n_neurons))

    # Perform the forward pass calculation
    def forward(self, inputs):
        # Calculate the output value from inputs, weights, and biases.
        # The output is the dot product of the inputs and weights, plus the biases.
        self.output = np.dot(inputs, self.weights) + self.biases
```
